# TensorFlow

## Components
1. Computation graph - Defines place holders and operations
2. Execution graph - Runs the graph

```python

# import libraries and modules needed for the code
import tensorflow as tf
import numpy as np

# Data
# A random matrix of size [3,5]
mat1 = np.random.rand(3,5)  
# A random matrix of size [5,2]
mat2 = np.random.rand(5,2) 


#Computation Graph
A = tf.placeholder(tf.float32, None, name='A')
B = tf.placeholder(tf.float32, None, name='B')
C = tf.matmul(A,B)


#Execution Graph
with tf.Session() as sess:
    result = sess.run(C, feed_dict={A: mat1, B:mat2})
    print(result)

## tf.session() creates a session to run the computation graph.
## A session allocates memory, manages cpu/gpu resources and executes operations in the graph.
## sess.run() runs the computation graph with the provided feed_dict to supply values for the placeholders A and B.
## result is a numpy arry containing the result of the matrix multiplication of mat1 and mat2.

###############################
### In Tensorflow 2.x  ###
##########################
import tensorflow as tf
import numpy as np

# Data
# A random matrix of size [3, 5]
mat1 = np.random.rand(3, 5)

# A random matrix of size [5, 2]
mat2 = np.random.rand(5, 2)

# Convert NumPy arrays to TensorFlow tensors
A = tf.constant(mat1, dtype=tf.float32)
B = tf.constant(mat2, dtype=tf.float32)

# Matrix multiplication
C = tf.matmul(A, B)

# Execution (automatic in TensorFlow 2.x)
print(C.numpy())

```

```python
import tensorflow as tf

from tensorflow import keras
A = np.random.rand(20, 500)
B = np.random.rand(500, 3000)

x = keras.backend.variable(A)
y = keras.backend.variable(B)
z = tf.linalg.matmul(x, y)
print(z.numpy())



## Code Explanation:
## Import the TensorFlow and NumPy libraries.   
## Create two random NumPy arrays A and B with specified shapes.
## Convert the NumPy arrays to TensorFlow variables x and y.
## Perform matrix multiplication using tf.linalg.matmul and store the result in z.
```
