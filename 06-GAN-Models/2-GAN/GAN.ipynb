## Generative Adversarial Networks (GAN)

- Generator - Try to create better images
- Discriminator - tries to detect differences.
- Nash equilibrium - both generator and discriminator competes untill - convergence. This happens when the discriminator has lost its ability to separate generated bills from real bills and essentially is randomly guessing.
- Both are neural Networks.
