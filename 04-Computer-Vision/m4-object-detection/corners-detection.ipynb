# Corner Detection

**Corner** : Rapid change in intensity where two edges meet.<br>
Derivatives are calculated to find the inensity of edge location.
- Gradients
- linear clusters
- Fitting ellipse to the distribution - Max and Mini Moment of Inertia (Ellipse Axes lenght)
- Compute second Moments for a region.
- For plane - Lambda 1 and Lambda2 are small
- For Edge - Lambda 1 is very large than Lambda 2 or vice versa
- For Corner - Lambda 1 and lambda 2 both are large.
- Harris Cornet Detection
